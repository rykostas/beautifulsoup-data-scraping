```python
#https://www.dataquest.io/blog/web-scraping-python-using-beautiful-soup/
#look at it again.to find all src bu ınside tıtle
import requests
from bs4 import BeautifulSoup
```

```python
pages=["https://www.carlogos.org/car-brands/"]
for i in range(1,9):
    temp=f"https://www.carlogos.org/car-brands/page-{i}.html"
    pages.append(temp)
    
pages
```

```python
allcars=[]

headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
allofthem=[]
for pag in pages:
    page = requests.get(url=pag, headers=headers)
    soup = BeautifulSoup(page.content, 'html.parser')
    for a in soup.select("ul.logo-list a img"):
        temp=a.get('alt').replace(' logo','')
        allcars.append(temp)
```

```python
allcars
```

```python
import pickle
with open('allcars.pkl', 'wb') as f:
    pickle.dump(allcars, f)
```

```python
with open('allcars.pkl', 'rb') as f:
    allcars = pickle.load(f)
```
