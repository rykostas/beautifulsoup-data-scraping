```python
header="https://www.gsmarena.com/"

allofthem=[]

for pag in pages:
    gezilecek=[pag]
    page = requests.get(pag)
    soup = BeautifulSoup(page.content, 'html.parser')
    for a in soup.select("div.nav-pages a"):
        temp=a.get("href")
        name=header+temp
        gezilecek.append(name)
        
    for gez in gezilecek: 
        page = requests.get(url=gez)
        soup = BeautifulSoup(page.content, 'html.parser')
        for a in soup.select("div.makers a strong span"):
            temp=pages[pag]+a.getText()
            allofthem.append(temp)
```
